```python
# Adjacency Matrix Representation
def find_max_degree_matrix(adj_matrix):
    max_degree = -1
    vertex_with_max_degree = -1

    # Calculate degree for each vertex
    for i in range(len(adj_matrix)):
        degree = sum(adj_matrix[i])  # sum of 1's in row i gives the degree
        if degree > max_degree:
            max_degree = degree
            vertex_with_max_degree = i

    return vertex_with_max_degree, max_degree


# Example Usage

# Adjacency Matrix (7 vertices, 16 edges)
adj_matrix = [
    [0, 1, 1, 1, 0, 0, 0],  # vertex 0
    [1, 0, 1, 1, 1, 0, 0],  # vertex 1
    [1, 1, 0, 0, 1, 1, 0],  # vertex 2
    [1, 1, 0, 0, 0, 1, 1],  # vertex 3
    [0, 1, 1, 0, 0, 1, 1],  # vertex 4
    [0, 0, 1, 1, 1, 0, 1],  # vertex 5
    [0, 0, 0, 1, 1, 1, 0]   # vertex 6
]

# Find vertex with maximum degree using Adjacency Matrix
vertex, degree = find_max_degree_matrix(adj_matrix)
print(f"Using Adjacency Matrix: Vertex {vertex} has the maximum degree of {degree}")
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()

# Add nodes and edges
edges = [
    (0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (1, 4),
    (2, 4), (2, 5), (3, 5), (3, 6), (4, 5), (4, 6),
    (5, 6)
]

G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold', edge_color='gray')
plt.title("Graph with 7 Vertices and 16 Edges")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Function to find the vertex with maximum degree (undirected graph)
def find_max_degree_matrix(adj_matrix):
    max_degree = -1
    vertex_with_max_degree = -1

    # Calculate degree for each vertex (undirected)
    for i in range(len(adj_matrix)):
        degree = sum(1 for j in range(len(adj_matrix)) if adj_matrix[i][j] != 0)  # count non-zero entries in row i
        if degree > max_degree:
            max_degree = degree
            vertex_with_max_degree = i

    return vertex_with_max_degree, max_degree

# Example: Edge-Weighted Adjacency Matrix (7 vertices, weighted edges)
adj_matrix = [
    [0, 4, 3, 2, 0, 0, 0],  # vertex 0
    [4, 0, 1, 3, 5, 0, 0],  # vertex 1
    [3, 1, 0, 6, 6, 7, 0],  # vertex 2
    [2, 3, 6, 0, 8, 8, 9],  # vertex 3
    [0, 5, 6, 8, 0, 2, 4],  # vertex 4
    [0, 0, 7, 8, 2, 0, 1],  # vertex 5
    [0, 0, 0, 9, 4, 1, 0]   # vertex 6
]

# Find vertex with maximum degree using Adjacency Matrix
vertex_degree, degree = find_max_degree_matrix(adj_matrix)
print(f"Using Adjacency Matrix: Vertex {vertex_degree} has the maximum degree of {degree}")

# Create an undirected graph from the adjacency matrix
G = nx.Graph()

# Add edges with weights from the adjacency matrix
for i in range(len(adj_matrix)):
    for j in range(i+1, len(adj_matrix)):  # Ensure undirected edges (i < j)
        if adj_matrix[i][j] != 0:  # Only add edges with non-zero weights
            G.add_edge(i, j, weight=adj_matrix[i][j])

# Draw the graph with edge weights
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(G)  # Layout for node positions
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold', edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(i, j): f'{G[i][j]["weight"]}' for i, j in G.edges()})
plt.title("Edge-Weighted Undirected Graph with 7 Vertices")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def find_max_degree_directed_matrix(adj_matrix):
    max_degree = -1
    vertex_with_max_degree = -1

    # Calculate the outgoing degree for each vertex
    for i in range(len(adj_matrix)):
        out_degree = sum(1 for j in range(len(adj_matrix)) if adj_matrix[i][j] != 0)  # count non-zero entries in row i
        if out_degree > max_degree:
            max_degree = out_degree
            vertex_with_max_degree = i

    return vertex_with_max_degree, max_degree

def find_vertex_max_weight_directed_matrix(adj_matrix):
    max_weight_sum = -1
    vertex_with_max_weight = -1

    # Calculate the sum of outgoing edge weights for each vertex
    for i in range(len(adj_matrix)):
        weight_sum = sum(adj_matrix[i][j] for j in range(len(adj_matrix)) if adj_matrix[i][j] != 0)  # sum weights of non-zero entries
        if weight_sum > max_weight_sum:
            max_weight_sum = weight_sum
            vertex_with_max_weight = i

    return vertex_with_max_weight, max_weight_sum

def plot_directed_graph(adj_matrix):
    G = nx.DiGraph()  # Directed graph

    # Add edges with weights
    n = len(adj_matrix)
    for i in range(n):
        for j in range(n):
            if adj_matrix[i][j] != 0:  # If there is an edge
                G.add_edge(i, j, weight=adj_matrix[i][j])

    # Plot the directed graph
    plt.figure(figsize=(8, 8))
    pos = nx.spring_layout(G)  # Layout for node positioning
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=12, font_weight='bold', edge_color='gray')

    # Draw edge labels (weights)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.title("Directed Graph with Edge Weights")
    plt.show()


# Directed, Edge-Weighted Adjacency Matrix (7 vertices, example directed edges with weights)
adj_matrix = [
    [0, 4, 3, 2, 0, 0, 0],  # vertex 0 has edges to vertices 1 (weight 4), 2 (weight 3), and 3 (weight 2)
    [4, 0, 1, 3, 5, 0, 0],  # vertex 1 has edges to vertices 0 (weight 4), 2 (weight 1), 3 (weight 3), and 4 (weight 5)
    [3, 1, 0, 6, 6, 7, 0],  # vertex 2 has edges to vertices 0 (weight 3), 1 (weight 1), 4 (weight 6), and 5 (weight 7)
    [2, 3, 6, 0, 8, 8, 9],  # vertex 3 has edges to vertices 0 (weight 2), 1 (weight 3), 5 (weight 8), and 6 (weight 9)
    [0, 5, 6, 8, 0, 2, 4],  # vertex 4 has edges to vertices 1 (weight 5), 2 (weight 6), 5 (weight 2), and 6 (weight 4)
    [0, 0, 7, 8, 2, 0, 1],  # vertex 5 has edges to vertices 2 (weight 7), 3 (weight 8), 4 (weight 2), and 6 (weight 1)
    [0, 0, 0, 9, 4, 1, 0]   # vertex 6 has edges to vertices 3 (weight 9), 4 (weight 4), and 5 (weight 1)
]

# Find vertex with maximum outgoing degree using Adjacency Matrix
vertex_degree, degree = find_max_degree_directed_matrix(adj_matrix)
print(f"Using Adjacency Matrix (Outdegree): Vertex {vertex_degree} has the maximum outgoing degree of {degree}")

# Find vertex with maximum sum of outgoing edge weights using Adjacency Matrix
vertex_weight, weight_sum = find_vertex_max_weight_directed_matrix(adj_matrix)
print(f"Using Adjacency Matrix (Outgoing Weight Sum): Vertex {vertex_weight} has the maximum outgoing edge weight sum of {weight_sum}")

# Plot the directed graph
plot_directed_graph(adj_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def convert_directed_to_undirected_matrix(adj_matrix):
    n = len(adj_matrix)  # Number of vertices
    undirected_matrix = np.zeros((n, n), dtype=int)  # Initialize an undirected matrix with zeros

    # Iterate through the directed adjacency matrix
    for i in range(n):
        for j in range(n):
            if adj_matrix[i][j] != 0:  # Check if there is an edge from i to j
                undirected_matrix[i][j] = adj_matrix[i][j]  # Add the directed edge to the undirected graph
                undirected_matrix[j][i] = adj_matrix[i][j]  # Also add the reverse edge (undirected nature)

    return undirected_matrix

# Given directed adjacency matrix with edge weights
adj_matrix = [
    [0, 4, 3, 2, 0, 0, 0],  # vertex 0 has edges to vertices 1 (weight 4), 2 (weight 3), and 3 (weight 2)
    [4, 0, 1, 3, 5, 0, 0],  # vertex 1 has edges to vertices 0 (weight 4), 2 (weight 1), 3 (weight 3), and 4 (weight 5)
    [3, 1, 0, 6, 6, 7, 0],  # vertex 2 has edges to vertices 0 (weight 3), 1 (weight 1), 4 (weight 6), and 5 (weight 7)
    [2, 3, 6, 0, 8, 8, 9],  # vertex 3 has edges to vertices 0 (weight 2), 1 (weight 3), 5 (weight 8), and 6 (weight 9)
    [0, 5, 6, 8, 0, 2, 4],  # vertex 4 has edges to vertices 1 (weight 5), 2 (weight 6), 5 (weight 2), and 6 (weight 4)
    [0, 0, 7, 8, 2, 0, 1],  # vertex 5 has edges to vertices 2 (weight 7), 3 (weight 8), 4 (weight 2), and 6 (weight 1)
    [0, 0, 0, 9, 4, 1, 0]   # vertex 6 has edges to vertices 3 (weight 9), 4 (weight 4), and 5 (weight 1)
]

# Convert directed adjacency matrix to undirected adjacency matrix
undirected_matrix = convert_directed_to_undirected_matrix(adj_matrix)

# Create a graph using NetworkX
G = nx.Graph()

# Add nodes and edges to the graph with weights
n = len(undirected_matrix)
for i in range(n):
    for j in range(i+1, n):  # Only consider the upper triangular matrix for undirected edges
        if undirected_matrix[i][j] != 0:
            G.add_edge(i, j, weight=undirected_matrix[i][j])

# Plot the graph
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=12, font_weight='bold', edge_color='gray')

# Draw edge labels (weights)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Undirected Graph with Edge Weights")
plt.show()
```
